```python
import librosa as lb
import os
import numpy as np
```

```python
if os.path.exists('feature')==False:
    os.mkdir('feature')
```

```python
dirs = 'gearbox'
output = 'feature'
for i in os.listdir(dirs):
    for x in os.listdir(os.path.join(dirs,i)):
        if os.path.exists(os.path.join(output,i))==False:
            os.mkdir(os.path.join(output,i))
        path = os.path.join(dirs,i,x)
        y, sr = lb.load(path, sr=None)
        energy = lb.feature.melspectrogram(y=y, sr=sr, power=1)
        result = lb.core.amplitude_to_db(energy)
        name = x.split('.')[0]
        filename = name+'.npy'
        out = os.path.join(output,i,filename)
        np.save(out, result)
```
