```python
import glob
import os
import numpy as np
import librosa as lb
import tensorflow as tf
import matplotlib.pyplot as plt
import sys
```

bikin yang test

```python
n_mels=128
n_frames=64
n_fft=1024
hop_length=512
power=2.0
n_hop_frames = 8
dims = n_mels * n_frames

data = []
for i in glob.glob("/content/drive/MyDrive/Tugas Akhir/dev_data/gearbox1/source_test/*.wav"):
    wav, sr = lb.load(i,sr=None) #sampling rate (none) agar pake sampling rate dari dataset bukan bawaaan dari librosa
    mel_spectrogram=lb.feature.melspectrogram(y=wav, sr=sr, n_fft=n_fft, hop_length=hop_length, n_mels=n_mels, power = power) #merubah file wav (1 dimensi) menjadi melspectogram (2 dimensi)
    log_mel_spectrogram = 20.0 / power * np.log10(np.maximum(mel_spectrogram, sys.float_info.epsilon))
    n_vectors = len(log_mel_spectrogram[0, :]) - n_frames + 1
    data.append(log_mel_spectrogram) # yang di append (disatuin) itu data melspec
```

```python
data = np.array(data)
print(data.shape)
```

```python
print(mel_spectrogram)
print(f'Log mel spectogram \n{log_mel_spectrogram}')
```

```python
vectors = np.zeros((n_vectors, dims))
print(vectors.shape)
vectors
```

```python
for t in range(n_frames):
  # print(t)
  vectors[:, n_mels * t : n_mels * (t + 1)] = log_mel_spectrogram[:, t : t + n_vectors].T
```

```python

vectors = vectors[: : n_hop_frames, :]
```

```python
print(log_mel_spectrogram.shape)
# log_mel_spectrogram

print(vectors.shape)
vectors
```

```python
for idx in range(len(glob.glob("/content/drive/MyDrive/Tugas Akhir/dev_data/gearbox1/source_test/*.wav"))):
  # data = np.zeros((len(glob.glob("/content/drive/MyDrive/Tugas Akhir/dev_data/gearbox/source_test/*.wav")) * vectors.shape[0], dims), float)
  vectors = vectors[: : n_hop_frames, :]
  if idx == 0:
    data = np.zeros((len(glob.glob("/content/drive/MyDrive/Tugas Akhir/dev_data/gearbox1/source_test/*.wav")) * vectors.shape[0], dims), float)
  data[vectors.shape[0] * idx : vectors.shape[0] * (idx + 1), :] = vectors
  # print(data.shape)
  print(vectors.shape)
  print(data.shape)
```
