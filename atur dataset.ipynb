```python
import glob
import os
import numpy as np
import librosa as lb
```

bikin yang test

```python
data = []
for i in glob.glob("/content/drive/MyDrive/Tugas Akhir/dev_data/gearbox1/source_test/*.wav"):
    wav, sr = lb.load(i,sr=None) #sampling rate (none) agar pake sampling rate dari dataset bukan bawaaan dari librosa
    melspec=lb.feature.melspectrogram(y=wav, sr=sr) #merubah file wav (1 dimensi) menjadi melspectogram (2 dimensi)
    data.append(melspec) # yang di append (disatuin) itu data melspec
```

```python
data = np.array(data)
print(data.shape)
```

```python
len(data)
```

```python
np.save ('x_test.npy', data) #jadi yang disimpan dalam numpy adalah data melspec dengan shape 128 x 313
```

bikin yang train

```python
nilai = []
for i in glob.glob("/content/drive/MyDrive/Tugas Akhir/dev_data/gearbox1/train/*.wav"):
    wav, sr = lb.load(i,sr=None) #sampling rate (none) agar pake sampling rate dari dataset bukan bawaaan dari librosa
    melspec=lb.feature.melspectrogram(y=wav, sr=sr) #merubah file wav (1 dimensi) menjadi melspectogram (2 dimensi)
    nilai.append(melspec) # yang di append (disatuin) itu data melspec
```

```python
nilai = np.array(nilai)
print(nilai.shape)
```

```python
np.save ('x_train.npy', nilai) #jadi yang disimpan dalam numpy adalah data melspec dengan shape 128 x 313
```

```python
directory = '/content/drive/MyDrive/Tugas Akhir/dev_data/gearbox1/source_test'

#Wadah append
labels = []
types={'00':0, '01':1, '02':2}
for i in os.listdir(directory):
  if (i.endswith('.wav')): #jika nama file berakhiran .wav maka akan lanjut jalan (kasus 1 folder banyak format)
     label=i.split('_')[1] #memisahkan simbol '_' dan diambil kata ke 4 yaitu (anomaly/normal)
     category=types[label]
     print(label, category)
     labels.append(category)
     #print(label)
```

```python
len(labels)
```

```python
np.save ('y_test.npy', labels)
```
