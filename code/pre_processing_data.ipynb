```python
import glob
import os
import numpy as np
import librosa as lb
from numpy import asarray
from numpy import savez_compressed
import sys
from librosa import display
import matplotlib.pyplot as plt
from tqdm import tqdm

from keras.preprocessing.image import load_img
import warnings
from keras.preprocessing.image import img_to_array
from keras.preprocessing.image import array_to_img
from tensorflow.image import resize
from tensorflow.keras.preprocessing.image import array_to_img
```

```python
hop_length = 512
n_fft = 1024
n_mels = 128
name_machine = ["bearing", "fan", "gearbox","slider", "ToyCar", "ToyTrain", "valve"]
section_name = ["00", "01", "02"]
domain_name = ["source", "target"]

for machine in name_machine:
    for sec in section_name:
        for dom in domain_name:
            data_pen = []
            for i in glob.glob(f"dev_data/{machine}/test/*.wav"):
                domain = i.split('_')[3]
                section = i.split('_')[2]
                if domain == dom and section == sec:
                    wav, sr = lb.load(i,sr=None) 
                    melspec = lb.feature.melspectrogram(y=wav, sr=sr, n_fft=n_fft, hop_length=hop_length, n_mels=n_mels, fmax=3500)
                    melspec = 20.0 / 2.0 * np.log10(np.maximum(melspec, sys.float_info.epsilon))
                    spectrogram = np.abs(melspec)
                    power_to_db = lb.power_to_db(spectrogram, ref=np.median)

                    plt.figure(figsize=(8, 7))
                    lb.display.specshow(power_to_db, sr=sr, x_axis='time', y_axis='mel', cmap='magma', hop_length=hop_length)
                    plt.axis('off')
                    plt.ylim([2048,4096])
                    plt.savefig(f"test/{i.split('/')[-1].split('.')[0]}.png", bbox_inches='tight', pad_inches=0, dpi=300, frameon='false')
                    plt.close()


                    img = load_img(f"test/{i.split('/')[-1].split('.')[0]}.png")

                    # convert the given image into  numpy array
                    img = resize(img,[64,128])
                    img_numpy_array = img_to_array(img)
                    img_numpy_array = img_numpy_array[:,:,1:2]
                    
#                     print(type(img_numpy_array))
                    data_pen.append(img_numpy_array)
            data_pen = np.array(data_pen)
#             print(data.dtype)
            np.save(f"bigan_dcase/data/test/x_test_{machine}_section_{sec}_{dom}.npy", data_pen)
            
            
            
            nilai_pen = []
            for i in glob.glob(f"dev_data/{machine}/train/*.wav"):
#                 domain = i.split('_')[3]
                section = i.split('_')[2]
                if section == sec:
                    wav, sr = lb.load(i,sr=None) 
                    melspec = lb.feature.melspectrogram(y=wav, sr=sr, n_fft=n_fft, hop_length=hop_length, n_mels=n_mels, fmax=3500)
                    melspec = 20.0 / 2.0 * np.log10(np.maximum(melspec, sys.float_info.epsilon))
                    spectrogram = np.abs(melspec)
                    power_to_db = lb.power_to_db(spectrogram, ref=np.median)

                    plt.figure(figsize=(8, 7))
                    lb.display.specshow(power_to_db, sr=sr, x_axis='time', y_axis='mel', cmap='magma', hop_length=hop_length)
                    plt.axis('off')
                    plt.ylim([2048,4096])
                    plt.savefig(f"train/{i.split('/')[-1].split('.')[0]}.png", bbox_inches='tight', pad_inches=0, dpi=300, frameon='false')
                    plt.close()


                    img = load_img(f"train/{i.split('/')[-1].split('.')[0]}.png")

                    # convert the given image into  numpy array
                    img = resize(img,[64,128])
                    img_numpy_array = img_to_array(img)
                    img_numpy_array = img_numpy_array[:,:,1:2]
                    
#                     print(type(img_numpy_array))
                    nilai_pen.append(img_numpy_array)
            nilai_pen = np.array(nilai_pen)
#             print(data.dtype)
            np.save(f"bigan_dcase/data/train/x_train_{machine}_section_{sec}.npy", nilai_pen)

        
```
